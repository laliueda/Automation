## Transferência

```python
# Salvar o arquivo PFD
py.hotkey('ctrl','tab')
    
def click_if_found(image_path, confidence=0.7):
    try:
        center = py.locateCenterOnScreen(image_path, confidence=confidence)
        if center:
            py.click(center.x, center.y)
            return True  # Indicate success
    except py.ImageNotFoundException:
        print(f"Image '{image_path}' not found.")
    return False  # Indicate failure
if click_if_found('exportar.png'):
    print('Clicked exportar')
else:
    click_if_found('recibo.png')
    print('Clicked recibo')
time.sleep(1.5)
py.moveTo(2775, 259)
py.click()
py.write('sisbajud')
py.press('enter')
py.press('left')
py.press('enter')
py.hotkey('ctrl','w')

# Abre as preferências no Eproc
py.keyDown('ctrl')
py.press('tab')
py.press('tab')
py.keyUp('ctrl')
py.press('pgdn')
pref = py.locateCenterOnScreen('preferencias.png', confidence = 0.7)
py.click(pref.x, pref.y)
juntInt = py.locateCenterOnScreen('ato intim.png', confidence = 0.6)
py.click(juntInt.x, juntInt.y)

# Intimação da fazenda e clica fora
py.moveTo(1390, 1541)
py.click()
py.press('pgdn')
time.sleep(1)
py.moveTo(319, 1325)
py.click()
py.press('tab')
py.write('60')
py.moveTo(1429, 1537)
py.click()
py.press('pgdn')

# Juntar o ato ordinatório
py.moveTo(442, 865)
py.click()
time.sleep(0.8)
py.press('tab')
py.press('enter')
py.keyDown('shift')
py.press('tab')
py.press('tab')
py.press('tab')
py.keyUp('shift')
py.write('sisbajud transferir')
time.sleep(0.8)
py.press('down')
py.press('enter')
py.keyDown('shift')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.keyUp('shift')
py.press('enter')
time.sleep(9)

# Junta o arquivo PDF
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('enter')
py.write('sisbajud.pdf')
py.press('enter')
py.press('tab')
py.press('tab')
py.write('sb')
time.sleep(1.5)
py.press('down')
py.press('enter')
py.press('tab')
kb.write('Transferência')
py.press('tab')
py.press('tab')
py.press('enter')
```
