```python
import pyautogui as py
import time
```

```python
py.PAUSE = 0.5
```

## Negativo - intimação da fazenda

```python
# Salvar o arquivo PFD
py.hotkey('ctrl','tab')
exp = py.locateCenterOnScreen('exportar.png', confidence = 0.7)
py.click(exp.x, exp.y)
time.sleep(1.5)
py.moveTo(2775, 259)
py.click()
py.write('sisbajud')
py.press('enter')
py.press('left')
py.press('enter')
py.hotkey('ctrl','w')

# Abre as preferências no Eproc
py.keyDown('ctrl')
py.press('tab')
py.press('tab')
py.keyUp('ctrl')
py.press('pgdn')
pref = py.locateCenterOnScreen('preferencias.png', confidence = 0.7)
py.click(pref.x, pref.y)
juntInt = py.locateCenterOnScreen('Junt int.png', confidence = 0.7)
py.click(juntInt.x, juntInt.y)

# Remove localizador e clica fora
py.moveTo(521, 770)
py.click()
py.moveTo(1245, 770)
py.click()
py.press('pgdn')

# Intimação da fazenda e clica fora
time.sleep(1)
py.moveTo(290, 1026)
py.click()
py.press('tab')
py.write('60')
py.moveTo(977, 1309)
py.click()

# Juntar o arquivo
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('tab')
py.press('enter')
py.write('sisbajud.pdf')
py.press('enter')
py.press('tab')
py.write('sb')
time.sleep(1.5)
py.press('down')
py.press('enter')
py.press('tab')
py.press('tab')
py.write('negativo')
py.press('tab')
py.press('enter')
```
